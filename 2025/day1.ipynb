```python
import os
from dotenv import load_dotenv
from aocd import get_data, submit

load_dotenv()

data = get_data(day=1, year=2025, session=os.getenv("AOC_SESSION"))

test_data = """L68
L30
R48
L5
R60
L55
L1
L99
R14
L82"""
```

```python
def change_lock_number(change: int, direction: str, current: int, max_number: int = 100):
   
    move = change if direction == "R" else -change
    raw = current + move

    if direction == "R":
        k0 = (max_number - current) % max_number
    else:
        k0 = current % max_number

    if k0 == 0:
        k0 = max_number

    if k0 > change:
        quotient = 0
    else:
        quotient = 1 + (change - k0) // max_number

    next_pos = raw % max_number
    return next_pos, quotient
  
current = 50 
password_a = 0
password_b = 0

for line in data.splitlines():
    if line == "":
        continue
    direction = line[0]
    change = int(line[1:])
    old = current
    current, quotient = change_lock_number(change=change, direction=direction, current=current)
    
    password_b += quotient
   
    if current == 0:
        password_a += 1
        
    
    
print(f"Password A: {password_a}, Password B: {password_b}")
```

```python
submit(password_a, day=1, year=2025, part="a", session=os.getenv("AOC_SESSION"))
```

```python
submit(password_b, day=1, year=2025, part="b", session=os.getenv("AOC_SESSION"))
```
