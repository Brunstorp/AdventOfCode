```python
import os
from dotenv import load_dotenv
from aocd import get_data, submit

import get_aoc_test_data

load_dotenv()

session = os.getenv("AOC_SESSION")

data = get_data(day=1, year=2025, session=session)

test_data = get_aoc_test_data(day=1, year=2025, session=session)
```

```python
def change_lock_number(change: int, direction: str, current: int, max_number: int = 100):
   
    move = change if direction == "R" else -change
    next = current + move
    quotient = abs(next // max_number)
    next = next % max_number
    return next, quotient
  
current = 50 
password_a = 0
password_b = 0

for line in data.splitlines():
    if line == "":
        continue
    direction = line[0]
    change = int(line[1:])
    old = current
    current, quotient = change_lock_number(change=change, direction=direction, current=current)
    
    password_b += quotient
   
    if current == 0:
        password_a += 1

password_b -= 1  #Adjust for overshoot since we miss one 0

print(f"Password A: {password_a}, Password B: {password_b}")
```

```python
submit(password_a, day=1, year=2025, part="a", session=os.getenv("AOC_SESSION"))
```

```python
submit(password_b, day=1, year=2025, part="b", session=os.getenv("AOC_SESSION"))
```
