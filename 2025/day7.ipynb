```python
import os
from dotenv import load_dotenv
from aocd import get_data, submit

from get_aoc_test_data import get_aoc_test_data
import numpy as np


load_dotenv()

session = os.getenv("AOC_SESSION")

day=7
year=2025

data = get_data(day=day, year=year, session=session)

test_data = get_aoc_test_data(day=day, year=year, session=session)
```

```python
problem = data
```

```python
matrix = np.array([list(line) for line in problem.splitlines()])
total_splits = 0

start = np.array([0, np.where(matrix[0] == 'S')[0][0]])
current_positions = set([tuple(start)]) # set of all current positions
next_positions = set() # set of all the next positions

for i in range(matrix.shape[0]):
    for current in current_positions:
        matrix[current[0], current[1]] = '|' # mark current position as visited
        next_position = np.array(current) + np.array([1, 0])
        
        if next_position[0] >= matrix.shape[0]:
            continue
        
        if matrix[next_position[0], next_position[1]] == '^':
            total_splits += 1
            next_positions.add(tuple(np.array(next_position) + np.array([0, -1]) ))
            next_positions.add(tuple(np.array(next_position) + np.array([0, 1]) ) )
        else:
            next_positions.add(tuple(next_position))
    
    current_positions = next_positions
    next_positions = set()
    
print(matrix)
    
print(total_splits)
```

```python
submit(total_splits, day=day, year=2025, part='a', session=os.getenv("AOC_SESSION"))
```
