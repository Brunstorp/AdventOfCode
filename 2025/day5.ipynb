```python
import os
from dotenv import load_dotenv
from aocd import get_data, submit

from get_aoc_test_data import get_aoc_test_data

load_dotenv()

session = os.getenv("AOC_SESSION")

data = get_data(day=5, year=2025, session=session)

test_data = get_aoc_test_data(day=5, year=2025, session=session)
```

```python
import re

range_pattern = r"^(\d+)-(\d+)$"

fresh_ingredients = set()

nbr_fresh_a = 0
    
for line in data.splitlines():
    m = re.match(range_pattern, line)
    if m:
        a, b = m.groups()
        fresh_ingredients.add( (int(a), int(b)) )
    
    else:
        if line.strip() == "":
            continue
        
        ingredient = int(line)
        for a, b in fresh_ingredients:
            if a <= ingredient <= b:
                nbr_fresh_a += 1
                break
            
print(nbr_fresh_a)
```

```python
submit(nbr_fresh_a, day=5, year=2025, part="a", session=os.getenv("AOC_SESSION"))
```

```python
# b

fresh_ingredients = set()

for line in data.splitlines():
    m = re.match(range_pattern, line)
    if m:
        a, b = m.groups()
        fresh_ingredients.add( (int(a), int(b)) )
    
    else:
        break

# apparently tuples are sorted lexicographically in python
sorted_ranges = sorted(fresh_ingredients)  

nbr_fresh_b = 0

# we know atleast start1 <= start2 and end1 <= end2
# they overlap only when start2 <= end1
for i in range(1, len(sorted_ranges) ):
    start1, end1 = sorted_ranges[i-1]
    start2, end2 = sorted_ranges[i]
    if start2 <= end1:
        # they overlap, create new range
        new_range = (start1, max(end1, end2))
        sorted_ranges[i] = new_range
        sorted_ranges[i-1] = None
        
for nbr_range in sorted_ranges:
    if nbr_range is not None:
        nbr_fresh_b += nbr_range[1] - nbr_range[0] + 1
        
print(nbr_fresh_b)
 
    
```

```python
submit(nbr_fresh_b, day=5, year=2025, part="b", session=os.getenv("AOC_SESSION"))
```
