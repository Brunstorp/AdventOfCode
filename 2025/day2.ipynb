```python
import os
from dotenv import load_dotenv
from aocd import get_data, submit

load_dotenv()

data = get_data(day=2, year=2025, session=os.getenv("AOC_SESSION"))
print(data)


test_data = '''11-22,95-115,998-1012,1188511880-1188511890,222220-222224,
1698522-1698528,446443-446449,38593856-38593862,565653-565659,
824824821-824824827,2121212118-2121212124'''
```

```python
import re

def get_number_invalid_a(number: str) -> bool:
    return len(number) % 2 == 0 and number[0 : len(number) // 2] == number[len(number) // 2 :]

def get_number_invalid_b(number: str) -> bool:
    same = any(
    all(number[i:i+j] == number[0:j] for i in range(0, len(number), j))
    for j in range(1, len(number)) if len(number) % j == 0
    )
    return same
            
ranges = re.findall(r"(\d+)-(\d+)", data)

id_sum_a = 0
id_sum_b = 0
for start, end in ranges:
    for number in range(int(start), int(end) + 1):
        if get_number_invalid_a(str(number)):
            id_sum_a += number
    
        if get_number_invalid_b(str(number)):
            id_sum_b += number
            
            
print(id_sum_a)
print(id_sum_b)
```

```python
submit(id_sum_a, day=2, year=2025, part="a", session=os.getenv("AOC_SESSION"))
```

```python
submit(id_sum_b, day=2, year=2025, part="b", session=os.getenv("AOC_SESSION"))
```
