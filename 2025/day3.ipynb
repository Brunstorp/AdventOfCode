```python
import os
from dotenv import load_dotenv
from aocd import get_data, submit, get_puzzle
from get_aoc_test_data import get_aoc_test_data

load_dotenv()

session = os.getenv("AOC_SESSION")

data = get_data(day=3, year=2025, session=session)

test_data = get_aoc_test_data(day=3, year=2025, session=session)
```

```python

def solution(number: str, digits: int = 12) -> int:
    numbers = [int(num) for num in number]
    N = len(numbers)
    drop = N - digits
    
    stack = []
    
    for num in numbers:
        while drop > 0 and len(stack) > 0 and stack[-1] < num:
            stack.pop()
            drop -= 1
        stack.append(num)  
                
    result_digits = stack[:digits]
    result = sum([num*10**(i) for i, num in enumerate(reversed(result_digits))])
    
    return result
            

sum_banks_a = 0
sum_banks_b = 0

for number in data.splitlines():
    sum_banks_a += solution(number, digits=2)
    sum_banks_b += solution(number, digits=12)
    
    
print(sum_banks_a)
print(sum_banks_b)
```

```python
submit(sum_banks_a, day=3, year=2025, part="a", session=os.getenv("AOC_SESSION"))
```

```python
submit(sum_banks_b, day=3, year=2025, part="b", session=os.getenv("AOC_SESSION"))
```
