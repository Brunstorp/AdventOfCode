```python
import os
from dotenv import load_dotenv
from aocd import get_data, submit

from get_aoc_test_data import get_aoc_test_data
import numpy as np


load_dotenv()

session = os.getenv("AOC_SESSION")

day=8
year=2025

data = get_data(day=day, year=year, session=session)

test_data = get_aoc_test_data(day=day, year=year, session=session)
```

```python
problem = data
print(len(problem.splitlines()), "lines in problem data")
```

```python
import re
coord_pattern = r'(-?\d+),(-?\d+),(-?\d+)'
coords = [tuple(map(int, re.findall(coord_pattern, line)[0])) for line in problem.splitlines()]

pairs_distances = []

for i in range(len(coords)):
    x1, y1, z1 = coords[i]
    for j in range(i + 1, len(coords)):
        x2, y2, z2 = coords[j]
        dist2 = (x1 - x2)**2 + (y1 - y2)**2 + (z1 - z2)**2
        pairs_distances.append((dist2, i, j))

pairs_distances.sort(key=lambda x: x[0])
```

```python
# now union find connected components

class DSU:
    def __init__(self, n):
        self.parent = list(range(n))
        self.size = [1]*n
        self.components = n
        

    def find(self, x):
        if self.parent[x] != x:
            self.parent[x] = self.find(self.parent[x])
        return self.parent[x]

    def union(self, a, b):
        ra, rb = self.find(a), self.find(b)
        if ra == rb:
            return False # merge didn't work, already connected
        
        # otherwise merge and perhaps switch ordering
        if self.size[ra] < self.size[rb]:
            ra, rb = rb, ra
        self.parent[rb] = ra
        self.size[ra] += self.size[rb]
        self.components -= 1
        return True
        
n = len(coords)
dsu_a = DSU(n)

K = 1000  # number of pairs to connect
for k in range(min(K, len(pairs_distances))):
    _, i, j = pairs_distances[k]
    dsu_a.union(i, j)
    
comps = {}
for i in range(n):
    r = dsu_a.find(i)
    comps[r] = comps.get(r, 0) + 1

sizes = sorted(comps.values(), reverse=True)
ans = sizes[0] * sizes[1] * sizes[2]
print(ans)
```

```python
n = len(coords)
dsu_b = DSU(n)
for dist2, i, j in pairs_distances:
    dsu_b.union(i, j)
    if dsu_b.components == 1:
        last = (i,j)
        break  
    
a = coords[last[0]]
b = coords[last[1]]
print("Last pair to connect:", a, b)
a_x = a[0]
b_x = b[0]

ans_b = a_x*b_x
```

```python
submit(ans, part="a", day=day, year=year, session=session)
```

```python
submit(ans_b, part="b", day=day, year=year, session=session)
```
